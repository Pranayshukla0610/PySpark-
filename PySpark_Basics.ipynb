```python
#Create Spark Session & Basic DataFrame

from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Basic1").getOrCreate()

data = [("Rohan",23),("Amit",25),("Riya",22)]
column = ['Name','Age']

df = spark.createDataFrame(data,column)
df.show()
```

```python
#Select Specific Columns

df.select("Name").show()
```

```python
#Filter Data

df.filter(df.Age > 23).show()
```

```python
#Add New Column

from pyspark.sql.functions import col

df.withColumn("Age_after_5_years", col("Age")+5).show()
```

```python
#Drop Column

df.drop("Age").show()
```

```python
#Group By & Aggregation

df.groupBy("Age").count().show()
```

```python
#Sort Data

df.orderBy(df.Age.desc()).show()
```

```python
#Handle Missing Values

data = [("Rohan",23),("Amit",None),("Riya",22)]
data_null = spark.createDataFrame(data, ['Name','Age'])

data_null.fillna({"Age":0}).show()
```

```python
#Apply Function (UDF)

from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

def age_category(age):
  return "Young" if age < 24 else "Adult"

udf_func = udf(age_category, StringType())

df.withColumn("Category",udf_func("Age")).show()
```
