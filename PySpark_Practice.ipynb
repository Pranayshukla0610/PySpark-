```python
#Create DataFrame from Dictionary

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("DataFrame").getOrCreate()

data = [{"name":"Parth","age":23},{"name":"Raman","age":21}]
df = spark.createDataFrame(data)
df.show()
```

```python
df.printSchema()
```

```python
df.show(1)
```

```python
df.columns
```

```python
df.withColumnRenamed("name","Name").show()
```

```python
#Filter Using SQL Syntax

df.filter("age > 22").show()
```

```python
df.filter((df.age > 22) & (df.name == "Parth")).show()
```

```python
from pyspark.sql.functions import lit

df.withColumn("Country",lit("India")).show()
```

```python
df.withColumn("age",df.age.cast("string")).printSchema()
```

```python
data = [("Amit",25),("Amit",25)]
df_dup = spark.createDataFrame(data,['Name','Age'])
df_dup.dropDuplicates().show()
```

```python
df.count()
```

```python
from pyspark.sql.functions import avg
df.groupBy().agg(avg("age")).show()
```

```python
#Group By Multiple Columns

data = [("Amit","IT",25), ("Rajdeep","HR",27),("Samaira","Admin",25)]
data = spark.createDataFrame(data,['Name','Department','Age'])
data.groupBy("Name","Department").count().show()
```

```python
df.orderBy("Age").show()
```

```python
df1 = spark.createDataFrame([(1,"Amit"),(2,"Riya")],['id','Name'])
df2 = spark.createDataFrame([(1,"IT"),(2,"HR")],['id','Department'])

df1.join(df2,"id").show()
```
